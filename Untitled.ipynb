```python
def write_into_file(res, day):
    res = re.findall(r'\d+', res)
    text = "{}; {},{}; {}.{};\n".format(day,res[0],res[1],res[2],res[3])
    path= "Sunrise"
    f = open(path + ".txt", 'a+')
    f.write(text)
```
